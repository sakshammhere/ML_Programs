```python
X=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=df['Price']
```

Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
X_train.columns
```

```python
df1=pd.DataFrame(lm.coef_,X_train.columns,columns=['Coeff'])
```

```python
df1
```

Predictions

```python
y_pred=lm.predict(X_test)  # only features are passed here, which the model hasnt seen(Testing features data)
```

```python
y_pred
```

```python
y_test
```

```python
plt.scatter(y_test,y_pred)
```

```python
sns.histplot((y_test-y_pred))    #plotting the diff of predicted - actual
```

Compute Metrics

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error=metrics.mean_squared_error(y_test,y_pred)
```

```python
mean_squared_error
```

```python
np.sqrt(mean_squared_error)   #root mean squared error
```
