```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('titanic_train.csv')
```

```python
train.head()
```

```python
train.info()
```

```python
train.describe()
```

```python
train['Age'].fillna(train['Age'].median(),inplace=True)
```

```python
train.info()
```

```python
train['Cabin'].unique()
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
train['Embarked'].fillna('Unknown',inplace=True)
```

```python
train['Embarked'].unique()
```

```python
train.drop('Ticket',axis=1,inplace=True)
```

```python
train.drop('Name',axis=1,inplace=True)
```

```python
train=pd.get_dummies(train,columns=['Sex','Embarked'],drop_first=True)
```

```python
bool_columns=train.select_dtypes(include='bool').columns
```

```python
train[bool_columns]=train[bool_columns].astype(int)
```

```python
train.head()
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train['Embarked_Unknown'].value_counts()
```

```python
train.drop('Embarked_Unknown',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train.columns
```

```python
X=train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male','Embarked_Q', 'Embarked_S']]
```

```python
y=train['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm=LogisticRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
y_pred=lm.predict(X_test)
```

```python
y_pred
```

```python
compare=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
compare
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```
